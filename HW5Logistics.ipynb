```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import numpy as np
 
df = pd.read_excel('/Users/connerreinhardt/Google Drive/Supply Chain Logistics/HW5P1.xlsx', sheet_name='Sheet1')
 
print("Column headings:")
print(df.columns)

print(df[:])
```

```python
##create a cost vector using euclidian distances
costvec = np.zeros(shape=(24,24))
for i in range(24):
        for j in range(24):
                costvec[i,j] = abs(df['x-Coordinate'][i] - df['x-Coordinate'][j]) + abs(df['y-Coordinate'][i] - df['y-Coordinate'][j])
```

```python
##implement nearest neighbor
visited = np.zeros(24)
visited[0] = 1

order = [0]

iteration = 0
while len([i for i, x in enumerate(visited==0) if x]) > 0:
    print(order)
    startpoint = order[-1]
    mincost = 999999999
    nextmove = 0
    for j in [i for i, x in enumerate(visited==0) if x]:
        if costvec[startpoint,j] < mincost:
            mincost = costvec[startpoint,j]
            #print("try: ",startpoint,"->",j," mincost = ", mincost)
            nextmove = j
    order.append(nextmove)
    print('Insert', nextmove, 'with cost of', mincost)
    visited[nextmove] = 1
    iteration = iteration+1

print(order)
print('return to start')
order.append(0)
print(order)
```

```python
##output nearest neighbor
move_cost = []
for i in range(1,len(order)):
    move_cost.append(costvec[order[i-1],order[i]])

print("Order: ", order)
print("Move Cost: ", move_cost)
print("Total Cost: ", sum(move_cost))
```

```python
##implement cheapest insertion
##see steps printed below
visited = np.zeros(24)
visited[0] = 1
order = [0, 0]
startpoint = 0
move_cost = []

while len([i for i, x in enumerate(visited==0) if x]) > 0:
    print(order)
    mincost = 99999999
    toinsert = 0
    indextoinsert = 0
    for a in range(len(order)-1):
        #print("")
        #print(order[a], "->", order[a+1],"-- Cost:", costvec[order[a],order[a+1]])
        for i in [i for i, x in enumerate(visited==0) if x]:
            if (costvec[order[a],i] + costvec[i,order[a+1]] - costvec[order[a],order[a+1]]) < mincost:
                mincost = (costvec[order[a],i] + costvec[i,order[a+1]] - costvec[order[a],order[a+1]])
                #print("Check: ", order[a], "->",  i ,"->", order[a+1])
                #print("New cost: ",costvec[order[a],i], " ", costvec[i,order[a+1]])
                #print("Net: ", mincost)
                toinsert = i
                indextoinsert = a+1
    print("Insert",toinsert,"at index", indextoinsert, 'with net cost increase of', mincost)    
    print("")
    order.insert(indextoinsert,toinsert)
    visited[toinsert] = 1
```

```python
##output from cheapest insertion
move_cost = []
for i in range(len(order)-1):
    move_cost.append(costvec[order[i],order[i+1]])

print("Order: ", order)
print("Move Cost: ", move_cost)
print("Total Cost: ", sum(move_cost))
```
